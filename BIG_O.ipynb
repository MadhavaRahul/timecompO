```python
def runmama(n):
    for a in range(n):
        print(a) #O(n)
```

```python
runmama(9)
```

```python
def do(l):
    for a in range(l):
        print(a)
    for b in range(l):
        print(b)
        #O(n+n)=O(n)/O(2n) drop const.
do(10)
```

```python
def mo(o):
    for i in range(o):
        for j in range(o):
            print(i,j)
mo(10)
#O(n^2)
```

```python
def muo(o):#drop non dom
    for i in range(o):
        for j in range(o):
            print(i,j)
    for k in range(o):
        print(k)
muo(10)
#O(n^2)
```

```python
def fn(a):
    return a+a
fn(10)#O(1)
```
